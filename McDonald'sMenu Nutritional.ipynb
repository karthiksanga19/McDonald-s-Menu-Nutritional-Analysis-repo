```python
# Plot the distribution of calorie counts
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=30, kde=True, color='blue')
plt.title('Distribution of Calories Across Menu Items')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of Total Fat
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Fat'], bins=30, kde=True, color='red')
plt.title('Distribution of Total Fat Across Menu Items')
plt.xlabel('Total Fat (g)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of Protein
plt.figure(figsize=(10, 6))
sns.histplot(df['Protein'], bins=30, kde=True, color='green')
plt.title('Distribution of Protein Across Menu Items')
plt.xlabel('Protein (g)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of Carbohydrates
plt.figure(figsize=(10, 6))
sns.histplot(df['Carbohydrates'], bins=30, kde=True, color='orange')
plt.title('Distribution of Carbohydrates Across Menu Items')
plt.xlabel('Carbohydrates (g)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot for Calories vs. Total Fat
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Fat', y='Calories', data=df, hue='Category', palette='viridis')
plt.title('Calories vs. Total Fat')
plt.xlabel('Total Fat (g)')
plt.ylabel('Calories')
plt.show()
```

```python
# Pair plot for multiple relationships
sns.pairplot(df[['Calories', 'Total Fat', 'Protein', 'Carbohydrates']], kind='scatter', diag_kind='kde')
plt.suptitle('Pair Plot for Nutritional Components')
plt.show()
```

```python
# Plot of Vitamin A, C, Calcium, Iron % Daily Values
plt.figure(figsize=(12, 8))
df[['Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']].mean().plot(kind='bar', color=['skyblue', 'green', 'orange', 'red'])
plt.title('Average % Daily Values for Vitamins and Minerals')
plt.ylabel('Average % Daily Value')
plt.show()
```

```python
# List of columns that are numeric and relevant for analysis
numeric_columns = [
    'Calories', 'Total Fat', 'Protein', 'Carbohydrates'
]

# Group data by category and calculate mean values for relevant columns
category_means = df.groupby('Category')[numeric_columns].mean()

# Plot bar charts for each nutritional characteristic
category_means.plot(kind='bar', figsize=(12, 8), colormap='viridis')
plt.title('Mean Nutritional Values by Food Category')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram for Calories
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=30, kde=True, color='blue')
plt.title('Distribution of Calories Across Menu Items')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for Calories by Category
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Calories', data=df, palette='Set3')
plt.title('Calories Distribution by Food Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for Total Fat by Category
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Total Fat', data=df, palette='Set2')
plt.title('Total Fat Distribution by Food Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for Protein by Category
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Protein', data=df, palette='Set1')
plt.title('Protein Distribution by Food Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for Carbohydrates by Category
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Carbohydrates', data=df, palette='coolwarm')
plt.title('Carbohydrates Distribution by Food Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a figure with subplots
plt.figure(figsize=(15, 12))

# Subplot for Calories
plt.subplot(2, 2, 1)
sns.barplot(x='Category', y='Calories', data=df, ci=None, palette='viridis')
plt.title('Average Calories by Category')
plt.xticks(rotation=45)
```

```python
# Find the menu item with the highest calorie count
highest_calorie_item = df.loc[df['Calories'].idxmax()]
print("Menu item with the highest calorie count:")
print(highest_calorie_item[['Category', 'Item', 'Calories']])
```

```python
# Find the menu item with the lowest calorie count
lowest_calorie_item = df.loc[df['Calories'].idxmin()]
print("\nMenu item with the lowest calorie count:")
print(lowest_calorie_item[['Category', 'Item', 'Calories']])
```

```python
# List of columns that are numeric and relevant for analysis
numeric_columns = [
    'Calories', 'Calories from Fat', 'Total Fat', 'Total Fat (% Daily Value)',
    'Saturated Fat', 'Saturated Fat (% Daily Value)', 'Trans Fat', 'Cholesterol',
    'Cholesterol (% Daily Value)', 'Sodium', 'Sodium (% Daily Value)', 'Carbohydrates',
    'Carbohydrates (% Daily Value)', 'Dietary Fiber', 'Dietary Fiber (% Daily Value)',
    'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
    'Calcium (% Daily Value)', 'Iron (% Daily Value)'
]

# Calculate the average nutritional content for each category
average_nutritional_content = df.groupby('Category')[numeric_columns].mean()

print("\nAverage nutritional content of popular menu categories:")
print(average_nutritional_content)
```

```python
# Plot the average nutritional content for each category
average_nutritional_content.plot(kind='bar', figsize=(15, 8), colormap='viridis')
plt.title('Average Nutritional Content by Food Category')
plt.ylabel('Mean Nutritional Value')
plt.xticks(rotation=45)
plt.show()
```
